```python
# Using concat to merge the DataFreames
all_data = pd.concat([dec2021_df, jan2022_df, 
                     feb2022_df, mar2022_df,
                     apr2022_df, may2022_df,
                     jun2022_df, jul2022_df,
                     aug2022_df, sep2022_df,
                     oct2022_df, nov2022_df])
all_data.info()
```

```python
# drop null values for the all_data DataFrame
all_data.dropna(axis=0, inplace=True)
```

```python
# Check for duplicated data
all_data.drop_duplicates(inplace=True)
```

```python
# Convert started_at and ended_at columns into datetime datatype
all_data["started_at"] = pd.to_datetime(all_data["started_at"])
```

```python
all_data["ended_at"] = pd.to_datetime(all_data["ended_at"])
```

```python
# create "trip_duration" column using "ended_at" - "started_at"
all_data["trip_duration"] = all_data["ended_at"] - all_data["started_at"]
all_data
```

```python
# Check the statistic of trip_duration columns
all_data["trip_duration"].describe()
```

```python
# Find out the rows that has negative value in the trip_duration column
negative_duration = all_data.loc[(all_data["trip_duration"] < "00:00:00")]
negative_duration
```

```python
# drop the rows that have negative value in the trip_duration column
all_data.drop(negative_duration.index, axis=0, inplace=True)
all_data["trip_duration"].describe()
```

```python
# convert trip_duration to seconds and then to datetime
# in order to show trip_duration correctly in Tableau
all_data["trip_duration"] = all_data["trip_duration"] / np.timedelta64(1, 's')
```

```python
# convert trip_duration to int
all_data["trip_duration"] = pd.to_datetime(all_data["trip_duration"], unit="s")
all_data
```

```python
# write the all_data DataFrame to csv file
all_data.to_csv("./Data/all_trips_data.csv", index=False)
```
