### Merge all the DataFrame into one DataFrame

```python
# Using concat to merge the DataFreames
all_data = pd.concat([dec2021_df, jan2022_df, 
                     feb2022_df, mar2022_df,
                     apr2022_df, may2022_df,
                     jun2022_df, jul2022_df,
                     aug2022_df, sep2022_df,
                     oct2022_df, nov2022_df])
all_data.info()
```

### Cleaning

```python
# drop null values for the all_data DataFrame
all_data.dropna(axis=0, inplace=True)
```

```python
# Check for duplicated data
all_data.drop_duplicates(inplace=True)
```

```python
# Convert started_at and ended_at columns into datetime datatype
all_data["started_at"] = pd.to_datetime(all_data["started_at"])
```

```python
all_data["ended_at"] = pd.to_datetime(all_data["ended_at"])
```

### Create "trip_duration" column

```python
# create "trip_duration" column using "ended_at" - "started_at"
all_data["trip_duration"] = all_data["ended_at"] - all_data["started_at"]
all_data
```

```python
# Check the statistic of trip_duration columns
all_data["trip_duration"].describe()
```

```python
# Find out the rows that has negative value in the trip_duration column
negative_duration = all_data.loc[(all_data["trip_duration"] < "00:00:00")]
negative_duration
```

```python
# drop the rows that have negative value in the trip_duration column
all_data.drop(negative_duration.index, axis=0, inplace=True)
all_data["trip_duration"].describe()
```

```python
# Convert the trip_duration to minutes
all_data["trip_duration"] = all_data["trip_duration"] / dt.timedelta(minutes=1)
all_data["trip_duration"] = all_data["trip_duration"].astype("int32")
```

### Create name_of_weekday column

```python
# Create week variable
week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Create name of weekday column from started_at column
all_data["name_of_weekday"] = all_data["started_at"].dt.day_name()

# Use Categorical method on the name_of_weekday column so it will be in order for later use
all_data["name_of_weekday"] = pd.Categorical(all_data["name_of_weekday"], categories=week, ordered=True)
all_data
```

### Create the month column

```python
# Create month variable
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

# Create the "month" column from started_at column
all_data["month"] = all_data["started_at"].dt.month_name()

# Use Categorical method on the month column so it will be in order for later use
all_data["month"] = pd.Categorical(all_data["month"], categories=months, ordered=True)
all_data
```

### Create the hour column

```python
# Create the hour column from started_at column
all_data["hour"] = all_data["started_at"].dt.hour
all_data
```

# **EDA**

```python
# Create variable to hold the counts of member and casual riders
rider_type_count = all_data.groupby(all_data["member_casual"])[["ride_id"]].count()
rider_type_count
```

```python
# Create a bar chart for the counts of member and casual riders
rider_type_count.plot.bar(title = "Counts of Rider Types",
                          xlabel = "Numbers of Rider",
                          ylabel = "Rider Type")
```

Annual members uses the service more frequently than casual riders, almost 1/3 more of usage.

```python
# Check how riders uses different types of bike
rideable_by_rider_type = all_data.groupby(["rideable_type", "member_casual"]).count()["ride_id"].unstack()
rideable_by_rider_type.plot(kind="bar",
                            title="Types of Bike Usage by User Types",
                            xlabel="Bike Type")
```

Classic bikes have the highest utilization by both types of users, especially annual members; following by electric bikes; docked bikes have the lowest rate, and it's primarly used by casual users, there is no data in the period of this analysis for annual members use docked bikes.

### Average Trip Duration by User Types

```python
# Get the average trip duration by user types
avg_duration_by_types = all_data.groupby(["member_casual"])["trip_duration"].mean()
avg_duration_by_types
```

```python
# Make a bar plot for average trip duration by user types
avg_duration_by_types.plot.bar(title="Average Trip Duration by User Types",
                               xlabel="Member Type",
                               ylabel="Bike Usage")
```

Casual members have longer average trip duration, this could be that casual members are using the service for leisure; whereas annual members have average trip duration around 12 minutes, this could be that annual members use the service more for commuting to work, home, etc.

### Bike Usage by Hour

```python
# get the hourly bike usage data by user types
hour_by_type = all_data.groupby(["hour", "member_casual"]).count()["ride_id"].unstack()
hour_by_type
```

```python
# create plot for hour_by_type
hour_by_type.plot(kind="bar",
                  title="Bike Usage by Hour",
                  xlabel="Hour",
                  ylabel="Bike Usage"
                  )
```

From this data we can see casual users' usage for the service increase as the day goes, and have a peak around 4-6pm, on the other hand, annual members have peaks from 6-8am and 4-6pm, which are the typical commuting hours, so this could be a further proof of that the annual members use the servie for commuting.

### Bike Usage by Days

```python
# get the weekly bike usage data by user types
week_by_type = all_data.groupby(["name_of_weekday", "member_casual"]).count()["ride_id"].unstack()
week_by_type
```

```python
# Create a bar plot for hour_by_type
week_by_type.plot.bar(title="Bike Usage by Hour",
                      xlabel="Day of the Week",
                      ylabel="Bike Usage")
```

As menteioned before, casual users use the service more for leisure, the plot also support that as casual users have peak of bike usage during the weekend; and annual members have higher usage during the week and less then the weekend, which also agian shows that the annual members use the service for commuting. 

Overall, annual members have much higher utilizaiton for the service than casual users.
